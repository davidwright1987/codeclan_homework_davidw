```python
5. Filter the data so we only take “Classic Espresso Drinks”, and save this in a 
new data frame called espresso_drinks.
```

```python
espresso_drinks = drinks_content.loc[drinks_content["Beverage_category"] == "Classic Espresso Drinks"]
espresso_drinks
```

```python
6. Group your espresso_drinks data frame by the type of beverage prep, and then find out the 
mean calories for each drink group.
```

```python
espresso_drinks.groupby("Beverage_prep")["Calories"].mean()
```

```python
7. Get the same grouped mean Calories values as above, but this time sorted in descending order.
```

```python
espresso_drinks.sort_values(espresso_drinks.groupby("Beverage_prep")["Calories"].mean())
```
