```python
import pandas as pd
```

```python
songs = pd.read_csv("data/songs.csv")
```

```python
songs.head()
```

```python
songs.describe()
```

```python
songs = pd.get_dummies(songs, drop_first = True)
songs.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# give me the target vector
y = songs["target"]
# give me everything in the datframe apart from the target vector
X = songs.drop(columns = "target")
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.score(X, y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
pd.DataFrame({
    "Variable": X.columns.values,
    "Coeff": model.coef_
})
```
